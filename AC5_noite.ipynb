2) Receba uma palavra e determine se ela é um palindrome.

```

from IPython.display import display, HTML, Javascript

html ='''





<html lang="pt">

<head>
     <meta charset="utf-8" />
    <script>
          
        function palin(str){
            var palavra = String(str.value);
              /*Converter para minusculas e retirar os espaços*/
            palavra=palavra.toLowerCase().replace(/\s/g,"");
              /* Converter para string*/
            fraserReverse=palavra.split("").reverse().toString();

            /*Verificar número total de letras*/
            
            for (var i = 0; i < ((fraserReverse.length)-1); i++) {
                fraserReverse=fraserReverse.replace(",","");
            };

            /*Comparar frases*/

            if(palavra==fraserReverse){                
                resultado="É Palindrome ";
            }
            else{                
                resultado="Não é Palindrome";
            }

            console.log(resultado)
        }

        function enviar() {
      
            var pala = document.getElementById("pala");
            palin(pala);

        }  

          
    </script>
        
</head>
        
    <body>
              
            
        <form onsubmit="javascript: return enviar()">
              
                <label for="pala">Palavra</label> <br>
                <input type="text" id="pala"> <br>
                
                <input type="submit" value="Enviar">
        
        </form>       
                   
        
    </body>
        
</html>



'''

display(HTML(html))
```

```
from IPython.display import display, HTML, Javascript

html ='''

function Exercicio7() {

              var num = document.getElementById("InputAlteraPar").value;
              let count = 0
              for (let i = 0; i < num.length; i++) {
                  if (parseInt(num[i]) % 2 == 0) {
                      count += 1
                  }
              }
              console.log(`Quantidade de números pares é: ${count}`)
          }

'''

display(HTML(html))
```
